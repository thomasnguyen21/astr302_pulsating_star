```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
#comment out if this does not work
%matplotlib notebook 
```

```python
def sin_wave(coef, period, offset, x):
    
    y = coef * np.sin(x * 2 * np.pi/(period)) + offset
    
    return y


def cos_wave(coef, period, offset, x):
    
    y = coef * np.cos(x * 2 * np.pi/(period)) + offset
    
    return y
```

```python
dt = .01 #seconds
time = 10 #seconds

time = np.arange(0,20, dt) #hours

coef1 = 2
period1 = 2.5
offset1 = 0

counts1 = sin_wave(coef1, period1, offset1, time)


coef2 = 2
period2 = 4
offset2 = 0

counts2 = sin_wave(coef2, period2, offset2, time)
```

```python
counts = counts1 + counts2
```

```python
plt.figure('Ideal Response')
plt.subplot(211)
plt.plot(time, counts1, 'ko')
plt.plot(time, counts2, 'ro')
plt.xlabel('Time (hours)')
plt.ylabel('Counts')

plt.subplot(212)
plt.plot(time, counts, 'bo')
plt.xlabel('Time (hours)')
plt.ylabel('Counts')
plt.show()
```

```python
yf = fft(counts)
n_points = len(yf)
xf = fftfreq(n_points, dt)



plt.figure('FFT')
#plt.plot(1/xf, yf)
plt.stem(1/xf, np.abs(yf), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlim([0,10])
plt.ylabel('FFT Power')
plt.xlabel('Period (hours)')
plt.show()
```

```python
plt.figure('FFT Hz')
#plt.plot(1/xf, yf)
plt.stem(xf*0.000277777778, np.abs(yf), 'b', \
         markerfmt=" ", basefmt="-b")
#plt.xlim([0,10])
plt.ylabel('FFT Power')
plt.xlabel('Frequency ')
plt.show()
```
